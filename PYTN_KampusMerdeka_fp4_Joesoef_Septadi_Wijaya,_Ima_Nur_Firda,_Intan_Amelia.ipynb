```python
def plot_boxplot(df, continuous_features):
    data = df[continuous_features].copy()
    fig, axes = plt.subplots(nrows=len(data.columns)//2, ncols=2,figsize=(15,40))
    fig.subplots_adjust(hspace=0.7)
    font = {'family': 'serif',
        'color':  'darkblue',
        'weight': 'normal',
        'size': 16,
        }

    # Generate distplot
    for ax, feature in zip(axes.flatten(), data.columns):
        sns.boxplot(data[feature],ax=ax)
        ax.set_title(f'Analysis of {feature}', fontdict=font)
    plt.show()
```

```python
plot_boxplot(df, continuous_features)
```

```python
def remove_outliers(df,outliers_features):
    # create copy of dataframe

    data = df.copy()

    for feature in data[outliers_features].columns:
        Q3 = data[feature].quantile(0.75)
        Q1 = data[feature].quantile(0.25)
        IQR = Q3 - Q1
        lower_limit = round(Q1 - 1.5 * IQR)
        upper_limit = round(Q3 + 1.5 * IQR)
        data.loc[data[feature]>= upper_limit,feature] = upper_limit
        data.loc[data[feature]<=lower_limit,feature] = lower_limit
#         data = data[(data[feature] < upper_limit) & (data[feature] > lower_limit)]
    return data
```

```python
df = remove_outliers(df,outliers_features)
```

```python
plot_boxplot(df, continuous_features)
```

#Eksplorasi Data

```python
def feat_plot(feature):
    plt.figure(figsize=(16, 6))
    plt.subplot(1, 3, 1)
    feature.plot(kind = 'hist')
    plt.title(f'{feature.name}')

    #mean = feature.describe().mean()
    plt.subplot(1, 3, 2)
    mu, sigma = scipy.stats.norm.fit(feature)
    sns.distplot(feature)
    #plt.legend({'--': mu, 'sigma': sigma})
    plt.axvline(mu, linestyle = '--', color = 'green', )
    plt.axvline(sigma, linestyle = '--', color = 'red')
    plt.title(f'{feature.name} distribution')
```

#Data Preprocessing

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
```

```python
# Normalizing the Data
normalized_df = normalize(df)
```

```python
# Converting the numpy array into a pandas DataFrame
normalized_df = pd.DataFrame(normalized_df)
normalized_df
```

```python
# Reducing the dimensions of the data
pca = PCA(n_components = 2)
pcadf = pca.fit_transform(normalized_df)
pcadf = pd.DataFrame(pcadf)
pcadf.columns = ['Principal Component 1', 'Principal Component 2']

pcadf.head(10)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
scatter = ax.scatter(pcadf['Principal Component 1'],
                     pcadf['Principal Component 2'],
                     alpha=0.6)
plt.title('Plotting data', fontsize = 20)
plt.xlabel('Principal Component 1', fontsize = 15)
plt.ylabel('Principal Component 2', fontsize = 15)
plt.legend(*scatter.legend_elements(), loc="best", title="Principal\nComponent")
ax.plot([])
ax.grid()
plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
sse = []
for i in range(1,11):
    kmeanModel = KMeans(n_clusters=i)
    kmeanModel.fit(pcadf)
    sse.append(kmeanModel.inertia_)
plt.figure(figsize=(15,6))
plt.plot(range(1,11), sse, 'bx-')
plt.xlabel('Number of Clusters (k)', fontsize = 15)
plt.ylabel('Sum of Squared Error', fontsize = 15)
plt.title('The Elbow Method', fontsize = 20)
plt.show()
```

Kita akan memilih 3 sebagai k(number of cluster)

```python
Kmean = KMeans(n_clusters=3)
Kmean.fit(pcadf)
```

```python
pcadf['labels'] = Kmean.labels_
```

```python
sns.scatterplot(x = 'Principal Component 1', y = 'Principal Component 2', data = pcadf, hue = 'labels')
```

```python
kmedoid = KMedoids(n_clusters = 3, random_state=1)
kmedoid.fit(pcadf)
pcadf['KMedoids_Labels'] = kmedoid.predict(pcadf)
```

```python
sns.scatterplot(x = 'Principal Component 1', y = 'Principal Component 2', data = pcadf, hue = 'KMedoids_Labels')
```

```python
dbscan = DBSCAN(eps=2,min_samples=6)
dbscan.fit(pcadf)
pcadf['DBSCAN_Labels'] = dbscan.labels_
```

```python
sns.scatterplot(x = 'Principal Component 1', y = 'Principal Component 2', data = pcadf, hue = 'DBSCAN_Labels')
```

```python
score = silhouette_score(pcadf, pcadf['labels'])
print("Silhouette score KMean is {})".format(score))
```

```python
score = silhouette_score(pcadf, pcadf['KMedoids_Labels'])
print("Silhouette score KMedoid is {})".format(score))
```

```python
score = silhouette_score(pcadf, pcadf['DBSCAN_Labels'])
print("Silhouette score DBSCAN is {})".format(score))
```

Algoritma terbaik yang digunakan pada dataset dengan menggunakan silhouete score adalah KMedoids dengan nilai 0.8585899755366915. Silhouete score mengukur seberapa baik setiap sampel dalam sebuah klaster ditugaskan ke klasternya sendiri dibandingkan dengan klaster lainnya. Ini berkisar dari -1 hingga 1, di mana skor mendekati 1 menunjukkan bahwa sampel dikelompokkan dengan baik, skor mendekati 0 menunjukkan cluster yang tumpang tindih atau ambigu, dan skor negatif menunjukkan bahwa sampel mungkin ditugaskan ke cluster yang salah.
